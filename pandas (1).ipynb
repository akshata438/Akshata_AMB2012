```python
import pandas as pd
import numpy as np

data=pd.DataFrame({
    "plot":['A','B','C','D','E'],
    "yield":[2500,np.nan,2800,np.nan,2600],
    "rainfall":[150,200,np.nan,180,170],
    "temperature":[28,30,29,np.nan,27]
  })
print("data with missing numbers")
print(data)
```

```python
print("detecting missing numbers")
print("is null?")
print(data.isnull())
print(data.isnull().sum())
```

```python
data_mean=data.copy()
print(data_mean)
```

```python
data_mean["yield"]=data_mean["yield"].fillna(data_mean["yield"].mean())
print(data_mean["yield"])
```

```python
data_median=data.copy()
print(data_median["rainfall"])
data_median["rainfall"]=data_median["rainfall"].fillna(data_median["rainfall"].median())
print(data_median["rainfall"])
```

```python
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method="ffill")
print(data_ffill)
```

```python
data_dropped=data.dropna()
print(data_dropped)
```
