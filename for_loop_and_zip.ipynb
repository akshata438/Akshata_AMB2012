```python
pesticides=["neem oil","chloropyriphos","melathion","imidacloprid","spinosad"]
effectiveness=[75,85,70,90,88]

max_effectiveness=max(zip(effectiveness,pesticides))
print(max_effectiveness)
```

```python
print("most effective pesticide is",max_effectiveness[1])
```

```python
least_effective=min(zip(pesticides,effectiveness))
print("least effective pesticide is:",least_effective[0])
```

```python
average_effectiveness=sum(effectiveness)/len(pesticides)
print("average effectiveness is",average_effectiveness,"%")
```

```python
for i in range(len(pesticides)):
 if effectiveness[i]>=average_effectiveness:
  print(pesticides[i])
```

```python
for i in range(len(pesticides)):
  if effectiveness[i]>80:
   print("pesticides effectiveness is more than 80 are",pesticides[i])
```
